```r
library(MASS)
data7 <- read.table('../../mvadata/data7-1.txt', header=T)
t(data7)
data7$group <- factor(data7$group)
```

```r
pairs(data7[2:4], main='Pairs Plot', pch=21, bg=c('red', 'blue'))[unclass(data7$group)]
```

```r
# 판별분석
data7.lda <- lda(group ~ ., data=data7)
data7.lda
```

```r
pred.lda <- predict(data7.lda, newdata=data7)
t(pred.lda)
```

```r
# 분류표 및 오분류율
confm.lda <- table(data7$group, pred.lda$class)
confm.lda
```

```r
prop.table(confm.lda, 1)
```

```r
#정분류율
ret <- sum(diag(prop.table(confm.lda)))
ret
#오분류율
1 - ret
```

```r
# 변수 선택하기
library(klaR)
data7.forward <- greedy.wilks(group ~ ., data=data7)
data7.forward
# 변수가 x3 하나 선택됨
```

```r
data7.fwd.lda <- lda(data7.forward$formula, data=data7)
data7.fwd.lda
```

```r
pred.fwd.lda <- predict(data7.fwd.lda, newdata=data7)
t(pred.fwd.lda)
```

```r
confm.fwd <- table(data7$group, pred.fwd.lda$class)
confm.fwd
```

```r
error <- 1 - sum(diag(prop.table(confm.fwd)))
error
#오분류율이 감소
```
