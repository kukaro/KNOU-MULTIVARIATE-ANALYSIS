```r
library(readxl)
```

```r
alcohol <- read_xlsx('../../mvadata/alcohol.xlsx')
t(alcohol)
```

```r
summary(alcohol)
```

```r
# 판별분석 실행하기
library(MASS)
```

```r
alcohol.lda <- lda(TYPE ~ . ,data=alcohol)
alcohol.lda
# 그룹의 수가 3개이므로 2개의 선형판별 함수가 출력됨
```

```r
# 분류하기
pred.lda <- predict(alcohol.lda, newdata=alcohol)
t(pred.lda)
```

```r
t(pred.lda$class)
```

```r
t(pred.lda$posterior)
```

```r
t(pred.lda$x)
```

```r
# 모형평가 : Wilks Lambda 및 분류표
indep_vars <- alcohol[,-1]
t(indep_vars)
```

```r
alcohol.man <- manova(as.matrix(indep_vars) ~ pred.lda$class)
alcohol.wilks <- summary(alcohol.man, test='Wilks')
alcohol.wilks
# Wilks의 값이 0에 가깝다
```

```r
confm.lda <- table(alcohol$TYPE, pred.lda$class)
confm.lda
```

```r
prop.table(confm.lda, 1)
error <- 1 - sum(diag(confm.lda))/sum(confm.lda)
error
```

```r
# 변수 선택하기
library(klaR)
alcohol.forward <- greedy.wilks(TYPE ~ ., data=alcohol, niveau=0.01)
alcohol.forward
```

```r
# 변수 선택 후 판별분석 시행하기
alcohol.fwd.lda <- lda(alcohol.forward$formula, data=alcohol)
alcohol.fwd.lda
```

```r
pred.fwd.lda <- predict(alcohol.fwd.lda, newdata=alcohol)
confm.fwd <-table(alcohol$TYPE, pred.fwd.lda$class)
confm.fwd
```

```r
error <- 1 - sum(diag(confm.fwd)) /  sum(confm.fwd)
error
```

```r
# 판별함수 산점도
library(dplyr)
alcohol$TYPE <- recode(alcohol$TYPE, typeA='A', typeB='B', typeC='C')
pred.alc <- predict(alcohol.fwd.lda, newdata=alcohol)
plot(pred.alc$x[,1], pred.alc$x[,2], pch=19)
text(pred.alc$x[,1], pred.alc$x[,2], alcohol$TYPE, cex=0.7, pos=4, col='red')
abline(v=-2, lty=2)
lines(c(-2, 3.7), c(-0.1, -0.1), lty=2)
```
